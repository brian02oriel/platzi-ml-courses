````python
import mlflow
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import logging


logging.basicConfig(level=logging.WARN)
logger = logging.getLogger(__name__)
````

Para ejecutar mlflow en local se ejecuta en consola el siguiente comando: ``` mlflow ui ```

````python
mlflow.set_tracking_uri("sqlite:////home/brian02oriel/Documents/Platzi Courses/ml-courses/ml-ops/mlflow.db")
````

````python
print(f"tracking URI: '{mlflow.get_tracking_uri()}'")
````

````python

mlflow.set_experiment("iris_experiment") #Identifica el nombre del experimento en el UI de MLflow
with mlflow.start_run(run_name = "example_1"): #Se especifica el nombre de la ejecucion en el UI de MLflow

    X, y = load_iris(return_X_y= True)
    params = {"C": 0.1, 
              "random_state": 42}
    mlflow.log_params(params) # Mantiene un log de hiperparametros
    lr = LogisticRegression(**params).fit(X, y)
    y_pred = lr.predict(X)
    mlflow.log_metric("accuracy", accuracy_score(y, y_pred)) # Mantiene un log de metricas
    mlflow.sklearn.log_model(lr, "model")
    print(f"default artifact location: '{mlflow.get_artifact_uri()}'")
````

````python
mlflow.end_run()
````
