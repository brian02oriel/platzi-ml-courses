```python
import stanza
stanza.download('es')  # Download Spanish models
```

## Notes:
- processors: define the pipeline in a string
- pos: assign tags to the corpus

```python
nlp = stanza.Pipeline('es', processors = 'tokenize, pos')  # Initialize the Spanish pipeline
doc = nlp(' yo soy una persona muy amable')  # Process the text
```

```python
for sentence in doc.sentences:
    for word in sentence.words:
        print(f'{word.text} - {word.upos}')  # Print the word and its part-of-speech tag
```
