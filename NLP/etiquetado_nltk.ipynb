## Note:
NLTK don't have pre-trained algorythms for spanish tagging; so, training a tagger is needed

```python
nltk.download('cess_esp')
from nltk.corpus import cess_esp as cess
from nltk import UnigramTagger as ut
from nltk import BigramTagger as bt
```

```python
cess_sents = cess.tagged_sents()
fraction = int(len(cess_sents) * 0.9) # To use 90% of the data for training
uni_tagger = ut(cess_sents[:fraction]) # Train UnigramTagger
uni_tagger.accuracy(cess_sents[fraction:]) # Represents the accuracy of the successful tag assignment for the corpus cess_esp
```

```python
uni_tagger.tag("Yo soy una persona muy amable".split(" "))
```

```python
fraction = int(len(cess_sents) * 0.9) # To use 90% of the data for training
bi_tagger = bt(cess_sents[:fraction]) # Train BiTagger
bi_tagger.accuracy(cess_sents[fraction:]) 
```

```python
bi_tagger.tag("Yo soy una persona muy amable".split(" "))
```

### Note: 
UnigramTagger performs better than BigramTagger in this use case
